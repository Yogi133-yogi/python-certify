```python
### delete the nan values from the data

movies.dropna(axis=0,inplace=True)
```

```python
movies
```

```python
for i,j in movies.iterrows():
    if (j['unit'])=="Billions":
        movies.at[i,'budget']=j['budget']*1000
    elif (j['unit'])=="Thousands":
        movies.at[i,'budget']=j['budget']/1000000
  
```
